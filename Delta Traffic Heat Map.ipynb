```python
#Working API to get lat-lng from Google
#Defining my lists to append data
lato = []
lngo = []

params = {
    "key":g_key
}

#API URL
base_url = "https://maps.googleapis.com/maps/api/geocode/json"

for index, row in muni_o.iterrows():

    # get restaurant type from df
    dirso = row['Directions']
    #end_point = row['End_travel']

    # add keyword to params dict
    params['address'] = dirso
    #params['destinations'] = end_point

    #time.sleep(300)
    
    # assemble url and make API request
    #Request
    response = requests.get(base_url, params = params).json()

    lato.append(response["results"][0]["geometry"]["location"]["lat"])
    lngo.append(response["results"][0]["geometry"]["location"]["lng"])
print(response)
```

```python
coordinates_o = list(zip(lato,lngo))
```

```python
conco = muni_o["id_via"].tolist()
```

```python
figure_layout = {
    'width': '800px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
figo = gmaps.figure(layout = figure_layout)
heat = gmaps.heatmap_layer(coordinates_o, weights=conco)
figo.add_layer(heat)
figo
```

```python
#fig = gmaps.figure(layout=figure_layout)
# Assign the marker layer to a variable
markers = gmaps.marker_layer(coordinates_o)
# Add the layer to the map
figo.add_layer(markers)
figo
```
